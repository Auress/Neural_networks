Обзор сети NETNet – улучшенный вариант SSD (Single Shot Object Detection)

Название статьи: NETNet: Neighbor Erasing and Transferring Network for Better Single Shot Object Detection

Ссылка: https://arxiv.org/pdf/2001.06690.pdf

В данной статье представлен алгоритм сети на основе SSD улучшающий ее точность при минимальном увеличении времени обработки изображения.  
Алгоритм тестировался на наборе данных MS COCO.  
Указывается на два недостатка классической SSD:  
1. Легко потерять мелкие объекты   
2. Крупная (выступающая) часть большого объекта детектируется как отдельный объект.  

Основной принцип действия данного алгоритма — это отделять признаки небольших объектов от признаков больших объектов и оставлять признаки небольших объектов на поверхностных слоях, а признаки крупных объектов переносить на более глубокие слои. Для это имеется два модуля: стирания (feature eras-ing module (i.e., NEM)) и переноса (feature transferring module (i.e., NTM))  

В отличие от других методов детектирования (Scale-agnostic detector, Multi-branch scale-aware detector, Scale-aware detector), а также алгоритма, где хорошо распознанные области стираются для итеративного уточнения результатов прогнозирования, здесь предлагается генерировать признаки, зависящие от масштаба объекта для SSD путем введения механизма стирания и передачи. Стертые признаки в поверхностных слоях далее переносятся для улучшения признаков в глубоких слоях. Алгоритм стремимся удалить признаки больших объектов из поверхностных слоев пирамиды и создать признаки для детектирования мелких объектов. Признаки больших объектов в поверхностных слоях переносятся для усиления признаков глубоких слоев.

В результате улучшается соотношение качества к время затратам, т.к. снижается количество ошибок False positive (SSD – 49.2 FP/10k, NETNet – 47.2 FP/10k) и ошибок False negative на мелких объектах (SSD – 60.8%, NETNet – 39,8%).
